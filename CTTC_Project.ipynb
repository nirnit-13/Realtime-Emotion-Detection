```python
# importing packages
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os # helps to interact with the operating system
import pickle # helps to zip/compress the image file to byte(binary) format
```

```python
# accessing the images present in "Emotion"
# getcwd() - fetching the current working directory
img_dir = os.path.join(os.getcwd(), 'Emotion')
img_dir
```

```python
# to check the directory where currently we are working
print(os.getcwd())
```

```python
# creating a path to store the pickle files inside the data folder
img_final = os.path.join(os.getcwd(), "data")
img_final
```

```python
# to store all the input(images) and output(labels) in different []
images = []
labels = []
```

```python
# performing image preprocessing and saving the files
# listdir() - to list all the files/folders present in the directory
for i in os.listdir(img_dir): # to access the individual emntion folders
    for j in os.listdir(os.path.join(img_dir,i)): # to access the actual images from each folder
        img_path = os.path.join(os.path.join(img_dir,i),j)
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        img = cv2.resize(img, (48,48))
        images.append(img)
        labels.append(i.split(' ')[0]) 
    print(i, '-----> Completed')
    
```

```python
set(labels)
```

```python
images = np.array(images)
labels = np.array(labels)
```

```python
# to plot the image after pre-processing
plt.imshow(images[1000], cmap='gray')
plt.show()
```

```python
# Pickling the images and labels files
with open(os.path.join(img_final,'images.p'),'wb') as f:
    pickle.dump(images,f)

with open(os.path.join(img_final, 'labels.p'), 'wb') as f:
    pickle.dump(labels,f)
```
